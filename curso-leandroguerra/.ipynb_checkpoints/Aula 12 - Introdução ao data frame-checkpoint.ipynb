```python
#Criando o seu data frame 

import pandas as pd

v1 = [1958, 1962,1970,1994,2002]
v2 = [1,1,2,3,5]
v3 = [10,100,1000,5, 25]

df = pd.DataFrame({'v1': v1,'v2': v2,'v3': v3})

df
```

```python
# Como ler os dados do data frame?

#Acessando as colunas
df[["v3"]]
```

```python
df[["v1","v3"]]
```

```python
#Acessando as linhas

df.iloc[1]
```

```python
# Acessando o valor da linha 2, coluna 3

df.iloc[1][2]

# ou 

df.iloc[1,2]
```
