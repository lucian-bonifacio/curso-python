```python
clientes_antes_3_meses = clientes[0:973]
clientes_antes_3_meses
```

```python
clientes_pos_3_meses = clientes[973 : ]
clientes_pos_3_meses
```

```python
subset_antes_3_meses = clientes_antes_3_meses.drop_duplicates(subset='Cliente')
subset_antes_3_meses
```

```python
subset_pos_3_meses = clientes_pos_3_meses.drop_duplicates(subset='Cliente')
subset_pos_3_meses
```

```python
#Agora com os dois dataframes separados por data e clientes, vou fazer a união dos dois com o objetivo de ver quais nomes
#tem em comum. Os nomes em comum são aqueles que compraram nos dois períodos. Os nomes que compraram somente nos primeiros
#2 meses irão sumir. 
clientes_juntos = pd.merge(subset_antes_3_meses, subset_pos_3_meses, on='Cliente')
clientes_juntos
```

```python
#Agora vou fazer uma união grosseira com as tabelas dos clientes que compraram após três meses e com os clientes que
#compraram nos 2 períodos. Após feito isso, os nomes duplicados, claro, são aqueles que compraram nos dois períodos,
#assim sendo, se eu excluílos me sobrarão apenas os nomes que compraram após os 3 meses // SE EU NÃO ESTIVER ERRADO KKK.

clientes_final = pd.concat([subset_pos_3_meses, clientes_juntos ])
clientes_final
```

```python
clientes_final = clientes_final.drop_duplicates(keep=False)
clientes_final
```
